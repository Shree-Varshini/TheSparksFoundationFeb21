```python
# import libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
import matplotlib as plt
#%inline matplotlib
import seaborn as sns
```

```python
#Load the dataset
df = pd.read_csv("C:/Users/rpshr/Downloads/Iris.csv")
```

<h3>Exploratory data analysis</h3>

```python
df.head()
```

```python
df.info()
```

```python
sns.pairplot(df,hue='Species')
```

<h3>Feature Selection</h3>

```python
X = df.iloc[:, [1,2,3,4]].values
y = df.iloc[:,-1].values
```

```python
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

<h3>Building Decision Tree Model</h3>

```python
clf = DecisionTreeClassifier()
clf = clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

<h4>This model has an accuracy of about 95% which is really good.</h4>

<h3>Visualizing the Decision Tree for better understanding</h3>

```python
from sklearn import tree
fn = ['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn = ['setosa','versicolor','virginica']

fig,axes = plt.pyplot.subplots(nrows = 1,ncols = 1,figsize = (4,4),dpi=300)
tree.plot_tree(clf,feature_names = fn, class_names = cn, filled=True)
fig.savefig('DTC.png')
```
